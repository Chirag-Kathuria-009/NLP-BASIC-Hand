```
import spacy
m = spacy.load("en_core_web_sm")
```

```
example = "CEO of master Card is Ajay Banga"
doc = m(example)

for ent in doc.ents:
  print(ent.text,ent.label_)
```

```
example2 = "Apple is looking at buying U.K. startup for $1 billion"
doc1 = m(example2)

for ent in doc1.ents:
  print(ent.text,ent,ent.label_)
```

```
# For mountains
example3 = "The Ganges, the Indus, the Brahmaputra, the Mekong, the Yangtze and the Yellow Rivers all originate in the Himalayas. "
doc2 = m(example3)
for ent in doc2.ents:
  print(ent.text,ent.label_)
```
